```python
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt

img = mpimg.imread('/content/dog.jpeg')
img
```

```python
def color2grayscale(vector):
    return np.max(vector) * 2 + np.min(vector) * 0.5
```

```python
grayscale_02 = np.apply_along_axis(color2grayscale, axis = 2, arr = img)
plt.imshow(grayscale_02, cmap = plt.get_cmap('gray'))
plt.show()
```
